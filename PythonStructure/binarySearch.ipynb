```python
#binary search

def binary_search(list, item):
    low = 0
    high = len(list)-1
    while high>=low:
        mid = (low+high)//2
        if list[mid] == item :
            return mid
        
        if list[mid] < item:
            low = mid+1
        else:
            high = mid-1
    return -1
```
